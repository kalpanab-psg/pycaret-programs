
## Lets learn Regression in pycaret

```python
#Downlaod house price dataset- boston from datasets
from pycaret.datasets import get_data
boston_df=get_data('boston')
```

```python
#import regression library alone instead of full classification package
from pycaret.regression import *
```

```python

lr_model=setup(data=boston_df, target='medv')
```

```python
#compare models of regression based on regression metrics
compare_models()
```

```python
#choosing a model for our dataset
cat_boost=create_model('catboost')
```

```python
# tune the parameters of the model
cat_model_parameters_tuned=tune_model('catboost',n_iter=50, optimize='mae')
```

```python
interpret_model(cat_model_parameters_tuned)
    
```

```python
#predict on holdout 30%
predict_model(cat_model_parameters_tuned)
```

```python
#save model
save_model(cat_model_parameters_tuned , 'catboost_boston')
```

```python
#load model at a later date

cat_model = load_model(model_name='catboost_boston')
```
